## 단계구분도

```python
import pandas as pd
import folium
import json
```

```python
df = pd.read_csv('자치구별 공원 현황.csv')
```

```python
df.head()
```

```python
with open('C:/Workspace/02.DataAnalysis/04.공공API_및_지도시각화/seoul-gu-simple.json','rt',encoding='UTF8') as json_files:
    geo_data=json.load(json_files)
```

```python
center = [37.541,126.986]
map = folium.Map(location=center, zoom_start=11, tiles='Stamen Toner')

folium.Choropleth(
    geo_data=geo_data, 
    data=df,  
    columns=('지역', '공원면적'), 
    key_on='feature.id', 
    fill_color='BuPu', 
    legend_name='공원면적', 
).add_to(map)
title = '<h3 align="center" style="font-size:20px">서울시 자치구별 공원면적</h3>'
map.get_root().html.add_child(folium.Element(title))
map
```

```python
park = pd.read_csv('C://Workspace/02.DataAnalysis/05.Crawling/서울공원 정리.csv')
```

```python
map = folium.Map(location=center, zoom_start=11, tiles='Stamen Toner')

folium.Choropleth(
    geo_data=geo_data,
    data=df,
    columns=('지역','공원면적'),
    key_on='feature.id',
    fill_color='BuPu',
    legend_name='공원면적'
).add_to(map)

for i in park.index:
    folium.CircleMarker([park.위도[i], park.경도[i]], 
                        radius=int(park.크기[i]),
                        tooltip=f"{park.공원명[i]}({park.면적[i]}㎡)",
                        color='green', fill_color='green').add_to(map)

title = '<h3 align="center" style="font-size:20px">서울시 자치구별 공원면적</h3>'
map.get_root().html.add_child(folium.Element(title))
map
```
