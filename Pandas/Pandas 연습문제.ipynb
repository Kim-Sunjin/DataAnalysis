```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
species_list = []
feature_list = []
mean_list = []
std_list = []
for species in iris.species.unique():
    for feature in iris.columns[:-1]:
        cal = iris[iris.species == species][feature]
        species_list.append(species)
        feature_list.append(feature)
        mean_list.append(np.round(cal.mean(),2))
        std_list.append(np.round(cal.std(),4))

df = pd.DataFrame({
    'species':species_list,'feature':feature_list,'mean':mean_list,'std':std_list
})
df
```

```python
s = iris[iris.species=='serosa']['sepal_length']
q1, q3 = s.quantile(.25), s.quantile(.75)
iqr = q3-q1
q1, q3, iqr
```

```python
lower = q1 - 1.5 *iqr
upper = q3 + 1.5 *iqr
s[(s < lower) | (s > upper)] = np.nan
s.isnull().sum()
```

```python
def get_new_stat(s):
    q1, q3 = s.quantile(.25), s.quantile(.75)
    iqr = q3-q1
    ower = q1 - 1.5 *iqr
    upper = q3 + 1.5 *iqr
    s[(s < lower) | (s > upper)] = np.nan
    outlier = s.isnull().sum() != 0
    return outlier, s.mean(), s.std()
```

```python
species_list = []
feature_list = []
mean_list = []
std_list = []
outlier_list, new_mean_list, new_std_list = [],[],[]
for species in iris.species.unique():
    for feature in iris.columns[:-1]:
        cal = iris[iris.species == species][feature]
        species_list.append(species)
        feature_list.append(feature)
        mean_list.append(np.round(cal.mean(),2))
        std_list.append(np.round(cal.std(),4))

        outlier, new_mean, new_std = get_new_stat(cal)
        outlier_list.append(outlier)
        new_mean_list.append(new_mean)
        new_std_list.append(new_std)

df = pd.DataFrame({
    'species':species_list,'feature':feature_list,'mean':mean_list,'std':std_list, 
    'outlier':outlier_list, 'new_mean': new_mean_list, 'new_std': new_std_list
})
df.set_index(['species','feature'], inplace=True)
df
```
