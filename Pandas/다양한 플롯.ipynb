```python
y = [2, 3, 1]
x = np.arange(len(y))
xlabel = ['가', '나', '다']
plt.title("Bar Chart")
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.xlabel("가나다")
plt.ylabel("빈도 수")
plt.show()
```

```python
np.random.seed(2021)

people = ['몽룡', '춘향', '방자', '향단']
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.title("Barh Chart")
plt.barh(y_pos, performance, xerr=error, alpha=0.4)
plt.yticks(y_pos, people)
plt.xlabel('x 라벨')
plt.show()
```

### 타이타닉 선실별 승객수

```python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.title("Pie Chart")
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```

### 스캐터 플롯 (산점도)
- 2차원 데이터 즉, 두 개의 실수 데이터 집합의 상관관계를 살펴보려면 scatter 명령으로 스캐터 플롯을 그린다. 스캐터 플롯의 점 하나의 위치는 데이터 하나의 x, y 값이다.

```python
np.random.seed(2021)
X = np.random.normal(0, 1, 100)
Y = np.random.normal(0, 1, 100)
plt.title("Scatter Plot")
plt.scatter(X, Y)
plt.grid()
plt.show()
```

#### 3차원 혹은 4차원 산점도

```python
N = 30
np.random.seed(2021)
x = np.random.rand(N)
y1 = np.random.rand(N)
y2 = np.random.rand(N)
y3 = np.pi * (15 * np.random.rand(N))**2
plt.title("Bubble Chart")
plt.scatter(x, y1, c=y2, s=y3)
plt.show()
```

### Anscombe's quartet

```python
ans = sns.load_dataset('anscombe')
ans.shape
```

```python
ans.head()
```

```python
ans.groupby('dataset').describe()
```

### Dataset에 대해서 산점도와 회귀선 그리기

```python
ds1 = ans[ans.dataset == 'I']
ds1
```

```python
plt.scatter(ds1.x,ds1.y)
plt.grid()
plt.plot([2,16],[4,11],c='r', lw=2)
plt.title('Dataset I')
plt.show()
```

#### 4개의 데이터셋 모두에 대해서 그림

```python
fig, axes = plt.subplots(2,2, figsize=(8,6))

for i, roman in enumerate(['I', 'II', 'III', 'IV']):
    ds = ans[ans.dataset == roman]
    ax = axes[i//2, i%2]
    ax.scatter(ds.x,ds.y)
    ax.grid()

plt.show()
```

```python
fig, axes = plt.subplots(2,2, figsize=(8,6),sharex=True,sharey=True)

for i, roman in enumerate(['I', 'II', 'III', 'IV']):
    ds = ans[ans.dataset == roman]
    ax = axes[i//2, i%2]
    ax.scatter(ds.x,ds.y)
    ax.plot([3,20],[4.5,13], c='r',lw=3) # y= 0.5 * x + 3
    ax.set_title(f'Dataset {roman}')
    ax.grid()

fig.suptitle('Ancombe\'s Quartet', fontsize = 16)
plt.show()
```

## 박스 플롯

```python
iris= sns.load_dataset('iris')
```

```python
setosa = iris[iris.species == 'setosa']
plt.boxplot([setosa.sepal_length, setosa.sepal_width, setosa.petal_length, setosa.petal_width],
    labels=['sepal_length','sepal_width','setosa.petal_length', 'setosa.petal_width'])

plt.title('Setosa 품종의 각 feature별 박스 플롯')
plt.show()
```

```python
iris.species.value_counts()
```

```python
fig, axes = plt.subplots(1,3, sharey=True)

for i, species in enumerate(iris.species.unique()):
    spi = iris[iris.species == species]
    ax = axes[i]
    ax.boxplot([spi.sepal_length, spi.sepal_width, spi.petal_length, spi.petal_width],labels=['sepal_length','sepal_width','setosa.petal_length', 'setosa.petal_width'])

fig.suptitle('Iris 모든 품종의 feature별 박스 플롯', fontsize=16, y=1.02)
plt.show()
```
