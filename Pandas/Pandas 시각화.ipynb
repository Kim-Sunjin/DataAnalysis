```python
import numpy as np
import matplotlib as mpl
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(2021)
df1 = pd.DataFrame(np.random.randn(100, 3),
                   index=pd.date_range('1/1/2018', periods=100),
                   columns=['A', 'B', 'C']).cumsum()
df1.tail()
```

```python
df1.plot()
plt.title("Pandas의 plot메소드 사용 예")
plt.xlabel('시간')
plt.ylabel("Data")
plt.grid()
plt.show
```

```python
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')

iris.sepal_length[:20].plot(kind='bar', rot=0,
        color = plt.get_cmap('hot')(np.linspace(0.8,0.2,20)))
plt.title("꽃받침의 길이 시각화")
plt.xlabel("Data")
plt.ylabel("꽃받침의 길이")
plt.show()
```

```python
iris.sepal_length[:20].sort_values().plot(kind='barh', rot=0,
        color = plt.get_cmap('Blues')(np.linspace(0.8,0.2,20)))
plt.title("꽃받침의 길이 시각화")
plt.xlabel("Data")
plt.ylabel("꽃받침의 길이")
plt.show()
```

```python
iris[:5].plot.bar(rot=0)
plt.title("Iris 데이터의 Bar Plot")
plt.xlabel("Data")
plt.ylabel("각 Feature의 값")
plt.ylim(0, 7)
plt.show()
```

```python
iris[:5].plot.bar(rot=0)
plt.ylim(0,7)
plt.show()
```

```python
df2 = iris.groupby(iris.species).mean()
df2.columns.name = 'feature' # column 이름을 주면 그래프 범례에 표시된다.
df2
```

```python
df2.plot.bar(rot=0)
plt.title('각 족의 Feature 평균')
plt.xlabel('평균')
plt.ylabel('종')
plt.ylim(0,8)
plt.show()
```

```python
df2.T
```

```python
df2.T.plot.bar(rot=0)
plt.title("각 Feature의 종 별 평균")
plt.xlabel("Feature")
plt.ylabel("평균")
plt.show()
```

## pie chart

```python
df3 = titanic['class'].value_counts()
df3.plot.pie(autopct= '%.f2%%')
plt.title('선실별 승객수 비율')
plt.axis('equal')
plt.show()
```

## Histogram

```python
iris.plot.hist()
plt.title('각 Feature 값들의 빈도수 Histogram')
plt.xlabel('데이터 값')
plt.grid()
plt.show()
```

## box plot

```python
iris.plot.box()
plt.title("각 Feature 값들의 빈도수에 대한 Box Plot")
plt.xlabel("Feature")
plt.ylabel("데이터 값")
plt.grid()
plt.show()
```

```python
iris.boxplot(by = 'species')
plt.tight_layout(pad=3, h_pad = 1)
plt.show()
```
