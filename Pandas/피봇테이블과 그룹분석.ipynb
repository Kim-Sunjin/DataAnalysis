## 그룹분석

```python
iris = sns.load_dataset('iris')
iris.groupby(iris.species).mean()
```

```python
iris.groupby(iris.species).agg(['mean','std'])
```

```python
np.random.seed(0)
df2 = pd.DataFrame({
    'key1': ['A', 'A', 'B', 'B', 'A'],
    'key2': ['one', 'two', 'one', 'two', 'one'],
    'data1': [1, 2, 3, 4, 5],
    'data2': [10, 20, 30, 40, 50]
})
df2
```

### agg() 메서드로 함수 적용

```python
def peak_to_peak_ratio(x):
    return x.max()/ x.min()
```

```python
iris.groupby(iris.species).agg(peak_to_peak_ratio)
```

```python
iris.groupby(iris.species).agg(lambda x: x.max()/ x.min())
```

```python
groups = iris.groupby(iris.species)
groups.groups
```

```python
iris.describe
```

```python
iris.groupby(iris.species).describe().T
```

```python
def top3_petal_length(df):
    return df.sort_values(by="petal_length", ascending=False)[:3]

iris.groupby(iris.species).apply(top3_petal_length)
```

```python
mpg = sns.load_dataset('mpg')
mpg.head()
```

```python
mpg['manufacturer'] = mpg.name.apply(lambda x: x.split()[0])
```

```python
mpg['model'] = mpg.name.apply(lambda x: ' '.join(x.split()[1:]))
```

```python
def q3cut(s):
    return pd.qcut(s, 3, labels=["소", "중", "대"]).astype(str)


iris["petal_length_class"] = iris.groupby(iris.species).petal_length.transform(q3cut)
iris[["petal_length", "petal_length_class"]].tail(10)
```

## tips dataset

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.
```

```python
tips['tips_pct'] = np.round(tips.tip / tips.total_bill * 100,2)
```

```python
tips.describe()
```

```python
tips.groupby('sex').count()
```

```python
tips.groupby('sex').size()
```

```python
tips.groupby('sex')[['tips_pct']].mean()
```

```python
tips.pivot_table('tips_pct', ['sex','smoker'])
```

### 연습문제 4.7.3

```python
tips.groupby('day')[['tips_pct']].mean()
```

```python
def peak_to_peak(x):
    return x.max() - x.min()


tips.groupby(["sex", "smoker"])[["tip"]].agg(peak_to_peak)
```

```python
tips.groupby(["sex", "smoker"]).agg(["mean", peak_to_peak])[["total_bill"]]
```

```python
tips.pivot_table(['tips_pct', 'size'], ['sex', 'day'], 'smoker')
```
